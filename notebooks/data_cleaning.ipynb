```python
# quick_detect.py

import pandas as pd
import re
from datetime import datetime

# Load your data


print(" Quick Column Shift Detection\n")

# 1. Find future dates in Last_Apply_Date
print("1 Checking for future dates...")
if 'Last_Apply_Date' in df.columns:
    def is_future_date(val):
        try:
            date = datetime.strptime(str(val).strip(), '%d %b %Y')
            return date > datetime.now()
        except:
            return False
    
    future_dates = df[df['Last_Apply_Date'].apply(is_future_date)]
    print(f"   Found {len(future_dates)} rows with future dates")
    if not future_dates.empty:
        print(f"   Row indices: {future_dates.index.tolist()[:20]}...")  # First 20

# 2. Find dates in Salary column
print("\n2 Checking for dates in Salary column...")
date_pattern = r'\d{1,2}\s+(Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)\s+\d{4}'
if 'Salary' in df.columns:
    dates_in_salary = df[df['Salary'].astype(str).str.contains(date_pattern, regex=True, na=False)]
    print(f"   Found {len(dates_in_salary)} rows with dates in Salary")
    if not dates_in_salary.empty:
        print(f"   Row indices: {dates_in_salary.index.tolist()[:20]}...")

# 3. Find "Other" duplication in Industry_Type
print("\n3 Checking for malformed Industry_Type...")
if 'Industry_Type' in df.columns:
    malformed = df[df['Industry_Type'].astype(str).str.contains(r'Other.*,.*Other', regex=True, na=False)]
    print(f"   Found {len(malformed)} rows with duplicate 'Other'")
    if not malformed.empty:
        print(f"   Row indices: {malformed.index.tolist()[:20]}...")

# 4. Combine all affected rows
all_affected = set()
if 'Last_Apply_Date' in df.columns:
    all_affected.update(future_dates.index.tolist())
if 'Salary' in df.columns:
    all_affected.update(dates_in_salary.index.tolist())
if 'Industry_Type' in df.columns:
    all_affected.update(malformed.index.tolist())

print(f"\n TOTAL AFFECTED ROWS: {len(all_affected)} out of {len(df)}")
print(f"   Percentage: {len(all_affected)/len(df)*100:.2f}%")

# Save affected rows
if all_affected:
    affected_df = df.loc[list(all_affected)]
    affected_df.to_csv("data/reports/problematic_rows.csv", index=True)
    print(f"\n Saved problematic rows to: data/reports/problematic_rows.csv")
```

```python
def is_description(description: str) -> bool:
    """
    Validate if string is a valid job description
    
    Expected: Long text (200-5000 chars), multiple sentences
    
    Args:
        description: String to validate
        
    Returns:
        True if valid description, False otherwise
    """
    if not isinstance(description, str):
        return False
    
    # Check if "NA" or empty
    if description.strip().upper() in ['NA', 'N/A', 'NOT AVAILABLE', '']:
        return False
    
    # Should NOT be a URL (catch misalignment)
    if description.startswith('http'):
        return False
    
    # Should NOT look like a timestamp
    if re.match(r'\d{4}-\d{2}-\d{2}', description[:10]):
        return False
    return True



def is_job_url(job_url: str) -> bool:
    """
    Validate if string is a valid job URL
    
    Expected: Naukri.com job URLs
    
    Args:
        job_url: String to validate
        
    Returns:
        True if valid URL, False otherwise
    """
    if not isinstance(job_url, str):
        return False
    
    # Remove whitespace
    job_url = job_url.strip()
    
    # Check basic URL structure
    url_pattern = r'^https?://(www\.)?naukri\.com/.+$'
    
    if not re.match(url_pattern, job_url, re.IGNORECASE):
        return False
    
    # Additional checks
    if len(job_url) < 30:  # URLs are typically longer
        return False
    
    if ' ' in job_url:  # URLs shouldn't have spaces
        return False
    
    return True
    


from datetime import datetime

```
