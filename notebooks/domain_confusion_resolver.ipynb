```python
def calculate_demand_score(data):
    weight = {
        "Job_Posting_Frequency": 0.4,
        "Trend_Direction": 0.3,
        "Skill_Diversity":0.15,
        "Experience_Fleibility":0.15
    }
    job_score_dict = {}
    max_freq = 0
    # calculating job posting frequency
    for domain in data.Job_Type.unique():
        freq = data[data["Job_Type"] == domain].shape[0]
        max_freq = max(max_freq , freq)
        job_score_dict[domain] = {"Frequency": freq}

    # normalizing against highest demand domain
    for domain in job_score_dict:
        score_part_relative_to_max_freq = job_score_dict[domain]["Frequency"]/max_freq
        score = round(score_part_relative_to_max_freq*10,2)
        job_score_dict[domain]["job_posting_score"] = score
   
    #    calculating trend_score
    current_month_data = data[data["Scraped_At_Cleaned"] >= '2026-01-01']
    previous_month_data = data[data["Scraped_At_Cleaned"] < '2026-01-01']

    for domain in data.Job_Type.unique():
        current_month_freq = current_month_data[current_month_data["Job_Type"] == domain].shape[0]
        previous_month_freq = previous_month_data[previous_month_data["Job_Type"] == domain].shape[0]
        if previous_month_freq == 0:
            trend_score = 10.0
        else:
            trend_change = (current_month_freq - previous_month_freq) / previous_month_freq
            trend_score = round(min(max(trend_change * 10, 0), 10),2)
        job_score_dict[domain]["trend_score"] = trend_score

    # skill diversity score
    unique_company_hiring_frequency = {}
    max_hiring_frequency = 0
    for domain in data.Job_Type.unique():
        frequency =  data[data["Job_Type"]==domain]["Company_Cleaned"].unique().shape[0]
        unique_company_hiring_frequency[domain] =frequency
        max_hiring_frequency = max(frequency, max_hiring_frequency)
    for domain in unique_company_hiring_frequency:
        relative_score = round((unique_company_hiring_frequency[domain]/max_hiring_frequency)*10,2)
        job_score_dict[domain]["Skill_diversity_score"]= relative_score
   
#    calculating experience flexibility score
    experience_score = {}
    # max_experience_frequency = data.shape[0]
    max_experience_frequency = df.shape[0]
    for domain in data.Job_Type.unique():
        freq=  data[data["Job_Type"]==domain].shape[0]
        experience_score[domain] = freq
    
    
    for domain in experience_score:
        relative_score = round((experience_score[domain]/max_experience_frequency)*10,2)
        job_score_dict[domain]["experience_flexibility_score"] = relative_score

    # calculating overall demand score
    for domain in job_score_dict:
        total_demand = job_score_dict[domain]['job_posting_score']*weight["Job_Posting_Frequency"] + job_score_dict[domain]["trend_score"]*weight["Trend_Direction"]+ job_score_dict[domain]["Skill_diversity_score"]*weight["Skill_Diversity"] + job_score_dict[domain]["experience_flexibility_score"]*weight['Experience_Fleibility']

        job_score_dict[domain]["demand_score"] = total_demand
    
    return job_score_dict


score = calculate_demand_score(new_df )
for domain in score:
    print(domain , score[domain]["demand_score"])
```

```python
def calculate_avaeage_saclary(data):
    data_grouped_by_domain = data.groupby("Job_Type")
    average_salary_dict = {}
    for domain, group in data_grouped_by_domain:
        # average_min_salary = group["Min_Salary"].mean()
        # average_max_salary = group["Max_Salary"].mean()
        filtered_group = group[(group["Min_Salary"] > 0) & (group["Max_Salary"] > 0)]
        average_min_salary = filtered_group["Min_Salary"].mean()
        average_max_salary = filtered_group["Max_Salary"].mean()
        average_salary_dict[domain] = {
            "Average_Min_Salary": round(average_min_salary, 2),
            "Average_Max_Salary": round(average_max_salary, 2)
        }
    return average_salary_dict
average_salary = calculate_avaeage_saclary(new_df)
print(average_salary)
    
```

#### calculate job_opening last 30 days

```python
def calculate_average_joob_openings(data):
    grouped = data.groupby("Job_Type")
    average_openings_dict = {}
    average_openings = 3 # assumed average openings per job posting
    for domain, group in grouped:
        # total_openings = group["Number_of_Openings"].sum()  # feature is not present in dataset
        total_openings = group.shape[0] * average_openings
        average_openings = total_openings / group.shape[0]
        average_openings_dict[domain] = round(average_openings, 2)
    return average_openings_dict


average_openings = calculate_average_joob_openings(new_df)
print(average_openings)
```
